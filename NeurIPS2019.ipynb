```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

page = requests.get('https://nips.cc/Conferences/2019/Schedule')
soup = BeautifulSoup(page.text, 'html.parser')
repo = soup.find(class_="col-xs-12 col-sm-9")
```

```python
tutorial_list = repo.find_all(class_='maincard narrower Tutorial')
print(len(tutorial_list))
```

### Scrape poster

```python
poster_list = repo.find_all(class_='maincard narrower Poster')
print(len(poster_list))
```

```python
df_poster_list = []
i = 0
for poster in poster_list:
    authors = poster.find(class_='maincardFooter').text.split(' · ')
    location_time = poster.find_all(class_='maincardHeader')[1].text.split(" @ ")
    location = location_time[0]
    time = location_time[1]
    title = poster.find(class_='maincardBody').text
    class_subclass = poster.find("a").text .split("\n                    ")[-1].split(" -- ")
    topic = class_subclass[0]
    subtopic = class_subclass[1]        
    ## get id and proceed to click through
    _poster_id_ = poster.get("id")[9:]
    _poster_details_page = requests.get("https://nips.cc/Conferences/2019/Schedule?showEvent="+_poster_id_)
    _poster_details_soup = BeautifulSoup(_poster_details_page.text, 'html.parser')
    speakers = _poster_details_soup.find_all(class_="btn btn-default")
    for speaker in speakers:
        _speaker_id_ = speaker.get("onclick")[13:-3]
        _speaker_page = requests.get("https://nips.cc/Conferences/2019/Schedule?showSpeaker="+_speaker_id_)
        _speaker_soup = BeautifulSoup(_speaker_page.text, 'html.parser')
        _speaker_repo = _speaker_soup.find(class_="maincard Remark col-sm-12")
        author = _speaker_repo.find("h3").text
        org = _speaker_repo.find("h4").text
        homepage = _speaker_repo.find("a")
        if homepage is not None:
            homepage = homepage.get("href")
        
        i += 1
        print(i, author)
        ## save to dict
        d = {"org":org, "author":author, "homepage":homepage, 
             "type":"Poster", "title":title, 
             "topic":topic, "subtopic":subtopic,
             "location":location, "time":time}
        df_poster_list.append(d)
    
df_poster = pd.DataFrame(df_poster_list)    
df_poster.shape
df_poster.head()
```

#### AutoML summary

```python
def change_org_name(row):
    text = row['org']
    if "Chinese Academy of Sciences" in text:
        text = "Chinese Academy of Sciences"
    elif "Washington" in text:
        text = "University of Washington"
    elif "Amazon" in text:
        text = "Amazon"
    elif "Google" in text:
        text = "Google"
    elif "Uber" in text:
        text = "Uber"
    elif "microsoft" in text.lower():
        text = "Microsoft"
    elif "South China University" in text:
        text = "South China University"
    row['org'] = text
    return row
```

```python
df_AutoML = df_poster[df_poster["subtopic"] == "AutoML"]
df_AutoML_clean = df_AutoML.apply(change_org_name, axis=1)
df_AutoML_clean
```

```python
df_AutoML_summary = pd.DataFrame(df_AutoML_clean.groupby(['org'])['org'].value_counts().sort_values(ascending=False)) 
# df_AutoML_summary.to_csv("df_AutoML_summary.csv")
df_AutoML_summary
```
